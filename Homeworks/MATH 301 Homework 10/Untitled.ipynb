```python
import numpy as np
```

```python
def f(x):
    return 2*x**2-4*x+31
```

```python
def isprime(num):
    for n in range(2,int(num**0.5)+1):
        if num%n==0:
            return False
    return True
```

```python
for i in range(1000000000):
    if(isprime(f(i))==0):
        print(i,",",end = '')
```
